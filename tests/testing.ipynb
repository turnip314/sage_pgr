```python
R = PolynomialRing(QQ, ['c', 'x', 'y'])
c, x, y = R.gens()
f1 = x**2 - 1 - c
f2 = x * y - 1
F = [f1, f2]
param_point = [0] #[QQ.random_element()-1/2 for _ in params]
test_param = [1/10]
parametric_geometric_resolution(F, 1, param_point, test_param)
```

```python
R = PolynomialRing(QQ, ['r', 'z', 'x', 'y'])
r, z, x, y = R.gens()
H = ((1 - x)^4 - x*y*(1 - x - x^2 + x^3 + x^2*y))
sys = [
    H,
    x*H.derivative(x)-r*y*H.derivative(y),
    x*y*z-1
]
parametric_geometric_resolution(sys, 1, param_point=[3], test_param=[3+1/10])
```

```python
R = PolynomialRing(QQ, ['x', 'y', 'c1', 'c2', 's1', 's2'])
x, y, c1, c2, s1, s2 = R.gens()
F = [
    c1+c2-x,
    s1+s2-y,
    c1^2+s1^2-1,
    c2^2+s2^2-1
]
parametric_geometric_resolution(F, 2, param_point=[1, 1], test_param=[1/2, 1/2])
```
